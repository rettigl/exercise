```python
from scipy.special import gamma
def chi2(x, n):
    return (1/2)**(n/2)*1/(gamma(n/2))*np.exp(-x/2)*x**(n/2-1)
```

```python
x = np.linspace(0, 30, 100)
plt.plot(x, chi2(x, 8))
```

```python
import pandas as pd
data = pd.read_csv("linear_regression.csv")
```

```python
data
```

```python
plt.errorbar(data["x"], data["y"], data["sy"])
```

```python
def weighted_sum(x, s):
    return (x/s**2).sum()/(1/s**2).sum()
```

```python
weighted_sum(data["x"], data["sy"])
```

```python
b = (weighted_sum(data["x"]*data["y"], data["sy"]) - weighted_sum(data["x"], data["sy"])*weighted_sum(data["y"], data["sy"]))/(weighted_sum(data["x"]**2, data["sy"])-weighted_sum(data["x"], data["sy"])**2)
print(b)
```

```python
a = weighted_sum(data["y"], data["sy"]) - b*weighted_sum(data["x"], data["sy"])
print(a)
```

```python
plt.errorbar(data["x"], data["y"], data["sy"])
plt.plot(data["x"], a+b*data["x"])
```

```python
var_a = weighted_sum(data["x"]**2, data["sy"])/(weighted_sum(data["x"]**2, data["sy"])-weighted_sum(data["x"], data["sy"])**2)/(1/data["sy"]**2).sum()
print(var_a)
```

```python
sigma_a = np.sqrt(var_a)
print(sigma_a)
```

```python
var_b = 1/(weighted_sum(data["x"]**2, data["sy"])-weighted_sum(data["x"], data["sy"])**2)/(1/data["sy"]**2).sum()
sigma_b = np.sqrt(var_b)
print(sigma_b)
```

```python
plt.errorbar(data["x"], data["y"], data["sy"])
plt.plot(data["x"], a+b*data["x"])
plt.plot(data.x, (a-sigma_a) + data.x*(b+sigma_b))
plt.plot(data.x, (a+sigma_a) + data.x*(b-sigma_b))
plt.plot(data.x, (a-sigma_a) + data.x*(b-sigma_b))
plt.plot(data.x, (a+sigma_a) + data.x*(b+sigma_b))
```

```python
def linear(x, a, b):
    return a + b*x
```

```python
from lmfit import Model
lmodel = Model(linear)
print(f"parameter names: {lmodel.param_names}")
print(f"independent parameters: {lmodel.independent_vars}")
```

```python
params = lmodel.make_params(a=1, b=1)
```

```python
result = lmodel.fit(data["y"], params, x=data["x"], weights=1/data["sy"])
print(result.fit_report())
```

```python
result.plot()
```

```python
def residual(x, y, sy, a, b):
    return ((y - (a + b*x))/sy)
```

```python
import xarray as xr
chi2map = xr.DataArray(data = np.zeros((100, 100)), coords={"a":np.linspace(-3.5, -2, 100), "b": np.linspace(2.8, 3.5, 100)})
```

```python
for i in range(len(chi2map.a)):
    for j in range(len(chi2map.b)):
        chi2map[i][j] = (residual(data["x"], data["y"], data["sy"], chi2map.a[i].data, chi2map.b[j].data)**2).sum()
```

```python
chi2map.loc[{"a":slice(-3.2, -2.2), "b":slice(2.9,3.4)}].plot(vmax=102)
chi2map.loc[{"a":slice(-3.2, -2.2), "b":slice(2.9,3.4)}].plot.contour(levels=[chi2map.min()+1])
```
